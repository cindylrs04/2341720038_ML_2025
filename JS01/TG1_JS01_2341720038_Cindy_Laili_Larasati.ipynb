Jawab:

1.	Privasi & Penyalahgunaan Data (Etika & Hukum)
Dampak:
Penggunaan AI sering melibatkan data pribadi dalam jumlah besar. Jika data tersebut dipakai tanpa izin atau tanpa perlindungan memadai, hal ini melanggar hak privasi. Contoh nyata adalah kasus Clearview AI, yang mengumpulkan miliaran foto wajah dari internet tanpa persetujuan pengguna untuk sistem pengenalan wajah. Kasus serupa juga terjadi pada Home Depot yang digugat karena menggunakan teknologi pengenalan wajah tanpa persetujuan pelanggan.
Referensi: The New York Times, Reuters, PC Gamer.

Solusi:
	Regulasi ketat: Penerapan aturan seperti GDPR (Eropa) atau BIPA (Illinois, AS) yang mewajibkan persetujuan eksplisit.
	Teknik privasi: Differential privacy, federated learning, dan enkripsi untuk melindungi data.
	Audit & transparansi: Kewajiban perusahaan melaporkan penggunaan data serta risiko etis dari model yang dikembangkan.
	Kesadaran pengguna: Edukasi publik agar memahami hak-hak mereka terkait data pribadi.

2.	Disinformasi & Deepfake
Dampak:
Teknologi deepfake memungkinkan pembuatan gambar, audio, dan video palsu yang sulit dibedakan dengan yang asli. Hal ini bisa digunakan untuk menyebarkan berita bohong, propaganda politik, hingga pemerasan. Misalnya, video deepfake tokoh politik yang disebarkan untuk memanipulasi opini publik menjelang pemilu. Dampak lanjutannya adalah hilangnya kepercayaan masyarakat terhadap media digital.
Referensi: BBC News, Brookings Institution, MIT Technology Review.
Solusi:
	Deteksi otomatis: Pengembangan sistem AI untuk mendeteksi deepfake secara cepat.
	Kebijakan hukum: Regulasi ketat terkait penyebaran konten manipulatif berbasis AI.
	Kolaborasi media: Platform media sosial wajib menandai atau menghapus konten terdeteksi palsu.
	Literasi digital: Edukasi masyarakat agar lebih kritis terhadap konten online.

3.	Bias & Diskriminasi dalam AI
Dampak:
AI dapat memperkuat bias yang ada di data latih. Misalnya, sistem rekrutmen berbasis AI yang cenderung memilih kandidat laki-laki karena data latihnya lebih banyak berisi pekerja pria. Akibatnya, muncul diskriminasi gender, ras, atau status sosial dalam pengambilan keputusan.
Referensi: Harvard Business Review, Nature Machine Intelligence, Wired.
Solusi:
	Dataset inklusif: Menggunakan data yang beragam dan seimbang untuk mengurangi bias.
	Audit independen: Penilaian eksternal terhadap sistem AI untuk mengidentifikasi diskriminasi.
	Transparansi algoritma: Membuka logika pengambilan keputusan AI kepada publik.
	Etika desain: Melibatkan pakar multidisiplin dalam perancangan sistem AI.
